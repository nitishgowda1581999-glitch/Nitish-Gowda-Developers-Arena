## Calculate Engagement Metrics

### Subtask:
Based on the available columns in the dataset, define and calculate appropriate engagement metrics (e.g., average time spent, frequency of use, interaction rates if available). Explain the rationale behind the chosen metrics.

**Rationale for Engagement Metrics:**

To understand social media engagement and its potential relation to addiction among students, we will focus on the following metrics available in the dataset:

1.  **Average Daily Usage Hours ('Avg_Daily_Usage_Hours'):** This directly quantifies the time spent on social media, serving as a primary indicator of engagement. Analyzing its central tendency and spread helps us understand typical usage patterns.
2.  **Most Used Platform ('Most_Used_Platform'):** Identifying the distribution of preferred platforms can reveal trends in where students spend most of their social media time, indirectly indicating platform-specific engagement.
3.  **Engagement Levels based on Daily Usage:** Segmenting students into 'low', 'medium', and 'high' usage categories provides a simplified but effective way to categorize and compare different engagement profiles within the student population.
4.  **Correlation between 'Avg_Daily_Usage_Hours' and 'Addicted_Score':** While 'Addicted_Score' itself is an outcome, its correlation with daily usage helps us understand if higher engagement (as measured by time spent) is associated with higher addiction levels, which is a crucial aspect of the task's objective.

**Reasoning**:
As per the subtask instructions, I will now calculate the mean, median, and standard deviation for the 'Avg_Daily_Usage_Hours' column to understand its central tendency and spread.



```python
print("\n--- Analysis of Avg_Daily_Usage_Hours ---")
print(f"Mean Daily Usage Hours: {df['Avg_Daily_Usage_Hours'].mean():.2f}")
print(f"Median Daily Usage Hours: {df['Avg_Daily_Usage_Hours'].median():.2f}")
print(f"Standard Deviation of Daily Usage Hours: {df['Avg_Daily_Usage_Hours'].std():.2f}")
```

**Reasoning**:
Next, as per the instructions, I will determine the distribution of the 'Most_Used_Platform' by calculating the percentage of students for each platform to identify which platforms attract the most engagement.



```python
print(
"\n--- Distribution of Most_Used_Platform ---"
)
platform_distribution = df['Most_Used_Platform'].value_counts(normalize=True) * 100
print(platform_distribution.round(2))
```

**Reasoning**:
As instructed, I will segment students into different engagement levels ('low', 'medium', 'high') based on their 'Avg_Daily_Usage_Hours' and then calculate the percentage of students in each category.



```python
print("\n--- Engagement Levels based on Avg_Daily_Usage_Hours ---")
def categorize_usage(hours):
    if hours < 3:
        return 'Low'
    elif 3 <= hours <= 6:
        return 'Medium'
    else:
        return 'High'

df['Engagement_Level'] = df['Avg_Daily_Usage_Hours'].apply(categorize_usage)
engagement_distribution = df['Engagement_Level'].value_counts(normalize=True) * 100
print(engagement_distribution.round(2))
```

**Reasoning**:
Finally, as per the subtask's instructions, I will calculate the Pearson correlation coefficient between 'Avg_Daily_Usage_Hours' and 'Addicted_Score' to understand the linear relationship between social media usage and addiction levels.



```python
correlation = df['Avg_Daily_Usage_Hours'].corr(df['Addicted_Score'])
print(f"\n--- Correlation between Avg_Daily_Usage_Hours and Addicted_Score ---")
print(f"Pearson Correlation Coefficient: {correlation:.2f}")
```

## Summary:

### Data Analysis Key Findings

*   **Overall Social Media Engagement:** Students spend an average of 4.92 hours daily on social media. The majority (73.62%) fall into a "Medium" engagement level (3-6 hours/day), while a significant portion (20.28%) are in the "High" engagement category (>6 hours/day). Instagram is the most popular platform, used by 35.32% of students.
*   **Strong Link Between Usage and Addiction:** There is a strong positive Pearson correlation coefficient of 0.83 between average daily social media usage hours and the addiction score, indicating that increased time spent on social media is highly associated with higher addiction levels.
*   **Impact on Academic Performance is Evident:** Students who reported that social media affects their academic performance also showed significantly higher average daily usage (5.54 hours vs. 3.80 hours), higher addiction scores (7.46 vs. 4.60), and notably lower average sleep hours (6.34 hours vs. 7.81 hours) and mental health scores (5.56 vs. 7.42) compared to those whose academics are not affected.
*   **Academic Level as a Risk Factor:** High School students exhibit the highest average daily usage (5.54 hours), addiction scores (8.04), and conflicts over social media (3.74), while also reporting the lowest average sleep hours (5.46 hours) and mental health scores (5.11). Conversely, Graduate students show the lowest usage and addiction scores and the highest sleep and mental health scores.
*   **Platform-Specific Impacts:**
    *   **Higher Impact Platforms:** WhatsApp, TikTok, and Snapchat are associated with higher average daily usage, higher addiction scores, more conflicts, and lower sleep hours and mental health scores. For instance, WhatsApp users reported the highest average daily usage (6.48 hours), highest addiction score (7.46), lowest sleep (5.87 hours), and lowest mental health score (5.54).
    *   **Lower Impact Platforms:** LINE and LinkedIn are linked to significantly lower average daily usage, lower addiction scores, fewer conflicts, and higher average sleep hours and mental health scores. LINE users, for example, reported the lowest addiction score (3.00) and the highest sleep hours (8.35 hours) and mental health scores (8.00).
*   **Gender Differences:** Females tend to have slightly higher average daily social media usage (5.01 hours vs. 4.83 hours), higher addiction scores (6.52 vs. 6.36), and more conflicts over social media (2.93 vs. 2.76) compared to males. Males, on average, report slightly more sleep and higher mental health scores.

### Insights or Next Steps

*   The clear correlation between social media usage and negative well-being indicators (addiction, reduced sleep, lower mental health scores, academic impact) suggests a critical need for awareness and intervention strategies, particularly targeting high-risk student groups like High Schoolers and those spending extended hours on specific platforms.
*   Further research into the specific features or interaction styles of platforms like WhatsApp, TikTok, and Snapchat that contribute to higher addiction and negative outcomes could inform the development of safer digital environments or responsible usage guidelines.
