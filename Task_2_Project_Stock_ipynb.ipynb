# Task
Analyze the AAL stock data from 'AAL.csv' by calculating and visualizing daily and cumulative returns, assessing its portfolio performance, and conducting a risk assessment using volatility, then summarize the findings.

## Load Stock Data

### Subtask:
Load the stock data from the provided CSV file 'AAL.csv' into a pandas DataFrame.


**Reasoning**:
Load the 'AAL.csv' file into a pandas DataFrame, display its first few rows and column information as instructed.



```python
import pandas as pd

df = pd.read_csv('/content/AAL.csv')
print("First 5 rows of the DataFrame:")
print(df.head())

print("\nDataFrame Info:")
df.info()
```

## Calculate Returns

### Subtask:
Calculate the daily and cumulative returns for the AAL stock.


**Reasoning**:
To calculate daily and cumulative returns, I need to first convert the 'Date' column to datetime objects and set it as the index. Then, I will calculate daily returns using the 'Adj Close' column and finally compute cumulative returns.



```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
df = df.set_index('Date')
df['Daily Returns'] = df['Adj Close'].pct_change()
df['Cumulative Returns'] = (1 + df['Daily Returns']).cumprod() - 1

print("First 5 rows with Daily and Cumulative Returns:")
print(df.head())
```

## Assess Portfolio Performance

### Subtask:
Analyze the performance of the AAL stock, treating it as a single-stock portfolio, by examining its cumulative returns over time.


**Reasoning**:
To analyze the performance of the AAL stock as a single-stock portfolio, I will plot the cumulative returns over time to visualize its growth.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Cumulative Returns'], label='AAL Cumulative Returns', color='blue')
plt.title('AAL Stock Cumulative Returns Over Time')
plt.xlabel('Date')
plt.ylabel('Cumulative Returns')
plt.grid(True)
plt.legend()
plt.show()
```

## Conduct Risk Assessment

### Subtask:
Calculate key risk metrics such as the volatility (standard deviation of daily returns) for the AAL stock.


**Reasoning**:
To calculate the volatility, I need to compute the standard deviation of the 'Daily Returns' column in the DataFrame and then print it with a descriptive label.



```python
volatility = df['Daily Returns'].std()
print(f"AAL Stock Volatility (Standard Deviation of Daily Returns): {volatility:.4f}")
```

## Visualize Analysis

### Subtask:
Generate plots to visualize the daily returns, cumulative returns, and a distribution of daily returns for risk assessment. Ensure all plots have appropriate legends.


**Reasoning**:
To visualize the daily returns, I will create a line plot showing 'Daily Returns' over time and a histogram to display the distribution of these daily returns, ensuring proper labels and legends for both plots.



```python
plt.figure(figsize=(14, 7))
plt.plot(df.index, df['Daily Returns'], label='AAL Daily Returns', color='green', alpha=0.7)
plt.title('AAL Stock Daily Returns Over Time')
plt.xlabel('Date')
plt.ylabel('Daily Returns')
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))
plt.hist(df['Daily Returns'].dropna(), bins=50, color='skyblue', edgecolor='black', label='Daily Returns Distribution')
plt.title('Distribution of AAL Stock Daily Returns')
plt.xlabel('Daily Returns')
plt.ylabel('Frequency')
plt.grid(True)
plt.legend()
plt.show()
```

**Reasoning**:
The subtask explicitly asks for visualization of cumulative returns, daily returns, and the distribution of daily returns. I have already visualized daily returns and their distribution in the previous step. Now I will visualize the cumulative returns over time to complete all visualization requirements for this subtask.



```python
plt.figure(figsize=(14, 7))
plt.plot(df.index, df['Cumulative Returns'], label='AAL Cumulative Returns', color='blue')
plt.title('AAL Stock Cumulative Returns Over Time')
plt.xlabel('Date')
plt.ylabel('Cumulative Returns')
plt.grid(True)
plt.legend()
plt.show()
```
